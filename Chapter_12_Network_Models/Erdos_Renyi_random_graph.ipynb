```python
## Endos Renyi Random network model 
import networkx as nx
import numpy as np 
from matplotlib import pyplot as plt

def random_graph(n,p):
    A = np.zeros((n,n))
    for i in range(n):
        for j in range(i+1,n):
            A[i,j] = np.random.rand()<p
            A[j,i] = A[i,j]

    return(A)
```

```python
n = 30
p = 0.1
A = random_graph(n,p)
G = nx.from_numpy_matrix(A)
nx.draw_circular(G, markersize = 0.01)
```

```python
n = 100
p = 0.1
A = random_graph(n,p)
degrees = np.sum(A,1)
plt.rcParams.update({'font.size': 18})
plt.hist(degrees)
plt.xlim([0,20])
plt.ylabel('$p(d)$')
plt.xlabel('$d$')
```
